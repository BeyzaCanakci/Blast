```python

from Bio.Blast import NCBIWWW
from Bio.Seq import Seq

protein_sequence = Seq("QFTDVDCSVSKECWSVCKDLFGVDRGKCMGKKCRCYQ")

search_in_database = NCBIWWW.qblast("blastp", "swissprot", protein_sequence)

from Bio.Blast import NCBIXML
blast_result = NCBIXML.read(search_in_database)

#Low e-value threshold is recommended for good alignment quality. This value is just to see all similar sequences. You can arrange the lower e-value threshold like 0.0000001 etc)
#Also, if you can see details about blast results, you can add extra information like this example
E_value_tresh = 0.004
count = 0
for blast_record in blast_result.alignments:
    for hsp in blast_record.hsps:
        if hsp.expect < E_value_tresh:
            count += 1
            print("-----------ALIGNMENT---------------")
            print(blast_record.title)
            print("Length:", blast_record.length)
            print("E-value:", hsp.expect)
            print("Max Score", hsp.bits)
            print("Total Score", hsp.score)
            print("Query Coverage", (hsp.align_length/len(protein_sequence)) * 100)
            print("Identities:", (hsp.identities / (hsp.align_length)) * 100)
            print(hsp.query)
            print(hsp.match)
            print(hsp.sbjct[0:37])

print(f"Blast process shows {count} similar sequences using swissprot database") 
```
